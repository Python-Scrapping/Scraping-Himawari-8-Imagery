```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://www.data.jma.go.jp/mscweb/data/himawari/list_r2w.html'
html = requests.get(url)
source = requests.get(url).text
soup = BeautifulSoup(source, 'lxml')
pd.set_option('display.max_colwidth', -1)
Himawari = []
Nama = []
```

```python
print(soup.prettify())
```

```python
for data in soup.find_all('a'):
    link = data['href']
    jam = link.split('/')[2]
    nama = jam.split('.')[0]
    src_hm = f'https://www.data.jma.go.jp/mscweb/data/himawari/'
    src = src_hm+link
    Himawari.append(src)
    Nama.append(nama)
```

```python
df = pd.DataFrame(Himawari, columns=['Alamat'])
df
```

```python
df_2 = pd.DataFrame(Nama, columns=['Nama File'])
df_2
```

```python
result = pd.concat([df, df_2], axis=1)
result.head()
```
